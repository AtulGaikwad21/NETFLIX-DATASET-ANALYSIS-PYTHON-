```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

# NETFLIX DATASET ANALYSIS

```python
# Load the dataset
netflix_data = '/kaggle/input/netflix-dataset-latest-2021/Netflix Dataset Latest 2021.xlsx'
netflix_data = pd.read_excel(netflix_data)

pd.DataFrame(netflix_data)


```

```python
# Show basic information about the dataset (rows, columns, data types)
print(netflix_data.info())
```

```python
pd.DataFrame(netflix_data.info())
```

```python
# Check the first few rows to confirm it's loaded
#netflix_data.head()
pd.DataFrame(netflix_data.head())
```

```python
# List of numeric columns where we want to fill missing values
numeric_columns = ['Hidden Gem Score', 'IMDb Score', 'Rotten Tomatoes Score', 
                   'Metacritic Score', 'Awards Received', 'Awards Nominated For', 
                   'Boxoffice', 'IMDb Votes']

# Fill missing values in numeric columns with the median value of each column
netflix_data[numeric_columns] = netflix_data[numeric_columns].fillna(netflix_data[numeric_columns].median())
pd.DataFrame(netflix_data[numeric_columns])
```

```python
# List of categorical (text) columns where we want to fill missing values
categorical_columns = ['Genre', 'Tags', 'Languages', 'Country Availability', 
                       'Director', 'Writer', 'Actors', 'View Rating', 'Production House']

# Fill missing values in categorical columns with 'Unknown'
netflix_data[categorical_columns] = netflix_data[categorical_columns].fillna('Unknown')
pd.DataFrame(netflix_data[numeric_columns] )
```

```python
# Display the updated information about the dataset (to check that missing values are handled).head()
pd.DataFrame(netflix_data[numeric_columns].head())
```

```python
# Display the updated information about the dataset (to check that missing values are handled)
print(netflix_data.info())
```

```python
# Get Basic Statistics
# Goal: Calculate basic statistics like the mean, count, and range for numeric columns.
# Get summary statistics for numeric columns
print(netflix_data.describe())
```

# DATA ANALYSIS WITH NUMPY AND PANDAS

```python
#Question 1: What is the average IMDb score of the movies in the dataset?
#Code:


# Calculate the average IMDb Score
average_imdb_score = netflix_data['IMDb Score'].mean()
print(f'Average IMDb Score: {average_imdb_score}')
```

```python
#Question 2: How many movies and how many series are there in the dataset?
#Code:


# Count the number of movies and series
movie_vs_series_count = netflix_data['Series or Movie'].value_counts()
#print(movie_vs_series_count)
pd.DataFrame(movie_vs_series_count)
#pd.DataFrame({'Question' :['movie_vs_series_count'],'TOTAL_COUNT':[movie_vs_series_count]})
```

```python
#Question 3: What is the total box office collection?
#Using Pandas to sum the box office earnings
total_box_office = netflix_data['Boxoffice'].sum()

# Creating a DataFrame for the result
pd.DataFrame({'Question': ['Total Box Office Collection'], 'Result': [total_box_office]})
```

```python
#Question 4: What is the correlation between IMDb score and box office?

# Using Pandas to find the correlation between IMDb score and box office
correlation = netflix_data[['IMDb Score', 'Boxoffice']].corr()

# Creating a DataFrame for the result
pd.DataFrame(correlation)
```

```python
#Question 5: What are the top 5 genres by frequency?

# Using Pandas to find the top 5 genres by count
top_genres = netflix_data['Genre'].value_counts().head(5)

# Creating a DataFrame for the result
pd.DataFrame(top_genres)
```

```python
#Question 6: What is the median Rotten Tomatoes score?

# Using Pandas to calculate the median Rotten Tomatoes score
median_rotten_tomatoes = netflix_data['Rotten Tomatoes Score'].median()

# Creating a DataFrame for the result
pd.DataFrame({'Question': ['Median Rotten Tomatoes Score'], 'Result': [median_rotten_tomatoes]})
```

```python
#question 7Find the Average IMDb Score
#Goal: Calculate the average (mean) IMDb score of all movies and series.


# Find the average IMDb score
average_imdb_score = netflix_data['IMDb Score'].mean()
print(f'Average IMDb Score: {average_imdb_score}')
```

```python
#question 8 : TOP DIRECTOR OF NETFLIX

# Grouping by Director and calculating the average IMDb score for each director
top_directors_by_score = netflix_data.groupby('Director')['IMDb Score'].mean().sort_values(ascending=False).head(10)

# Displaying the top 10 directors by average IMDb score
pd.DataFrame(top_directors_by_score)
```

```python
# Sorting the dataset by IMDb Score to get the top movies
top_movies_by_imdb = netflix_data[['Title', 'IMDb Score']].sort_values(by='IMDb Score', ascending=False).head(10)

# Displaying the top 10 movies by IMDb score
pd.DataFrame(top_movies_by_imdb)
```

```python
#QUESTION 10 :Top countries analysis?
# Splitting the 'Country Availability' column by multiple countries (in case content is available in multiple regions)
netflix_data['Country List'] = netflix_data['Country Availability'].str.split(',')

# Exploding the list into multiple rows (so each country has its own row)
country_data = netflix_data.explode('Country List')

# Counting the number of releases per country
top_countries = country_data['Country List'].value_counts().head(10)

# Displaying the top 10 countries with the most Netflix releases
pd.DataFrame(top_countries).reset_index().rename(columns={'index': 'Country', 'Country List': 'Number of Releases'})
```

# DATA ANALYSIS WITH matplotlib.pyplot

```python
#Question 1: What is the distribution of IMDb scores?

import matplotlib.pyplot as plt

# Plotting the distribution of IMDb Scores
plt.hist(netflix_data['IMDb Score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of IMDb Scores')
plt.xlabel('IMDb Score')
plt.ylabel('Frequency')
plt.show()
```

```python
#Question 2: How many movies vs series are there?

# Plotting the number of Movies vs Series
movie_vs_series = netflix_data['Series or Movie'].value_counts()
plt.bar(movie_vs_series.index, movie_vs_series.values, color=['blue', 'orange'])
plt.title('Number of Movies vs Series')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
#Question 4: What is the trend of Netflix releases over the years?

# Plotting the trend of Netflix releases
netflix_data['Year'] = netflix_data['Netflix Release Date'].dt.year
release_trend = netflix_data['Year'].value_counts().sort_index()

plt.plot(release_trend.index, release_trend.values, marker='o', color='purple')
plt.title('Trend of Netflix Releases Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.grid(True)
plt.show()
```

```python
#QUESTION 4 : Compare Netflix releases yearly .
# Extract the year from the Netflix release date
netflix_data['Year'] = netflix_data['Netflix Release Date'].dt.year

# Count the number of Netflix releases per year
yearly_releases = netflix_data['Year'].value_counts().sort_index()

# Plotting the number of Netflix releases per year using a bar graph
plt.figure(figsize=(10,6))
plt.bar(yearly_releases.index, yearly_releases.values, color='red')

plt.title('Number of Netflix Releases by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Releases', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

# DATA ANALYSIS With Using Matplotlib and Seaborn

```python
#Question 1: What is the relationship between IMDb score and Box Office earnings?

import seaborn as sns

# Plotting a scatter plot for IMDb Score vs Box Office
plt.figure(figsize=(8,6))
sns.scatterplot(x='IMDb Score', y='Boxoffice', data=netflix_data, color='blue')
plt.title('IMDb Score vs Box Office Earnings')
plt.xlabel('IMDb Score')
plt.ylabel('Box Office Earnings')
plt.grid(True)
plt.show()
```

```python
# Question 2: What is the distribution of awards received by IMDb score?

# Plotting a scatter plot with regression line for IMDb Score vs Awards Received
plt.figure(figsize=(8,6))
sns.regplot(x='IMDb Score', y='Awards Received', data=netflix_data, scatter_kws={'color':'green'}, line_kws={'color':'red'})
plt.title('IMDb Score vs Awards Received')
plt.xlabel('IMDb Score')
plt.ylabel('Awards Received')
plt.show()
```

```python
#Question 3: Top genres by IMDb?


# Grouping by Genre and calculating the average IMDb score for each genre
top_genres_by_imdb = netflix_data.groupby('Genre')['IMDb Score'].mean().sort_values(ascending=False).head(10)

# Plotting the top 10 genres by average IMDb score using Matplotlib and Seaborn
plt.figure(figsize=(10,6))
sns.barplot(x=top_genres_by_imdb.index, y=top_genres_by_imdb.values, palette="Blues_d")

plt.title('Top 10 Genres by Average IMDb Score', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Average IMDb Score', fontsize=12)
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```
